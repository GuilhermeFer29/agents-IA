MarkDownHeaderTextSplitter

```python
markdown_example = '''# Título do Markdown de exemplo
## Capítulo 1
Texto capítulo 1 e mais e mais texto.
Continuamos no capítulo 1!
## Capítulo 2
Texto capítulo 2 e mais e mais texto.
Continuamos no capítulo 2!
## Capítulo 3
### Seção 3.1
Texto capítulo 3 e mais e mais texto.
Continuamos no capítulo 3!
'''
```

```python
from langchain_text_splitters import MarkdownHeaderTextSplitter

headr_to_split_on=[
    ("#", "Header 1"),
    ("##", "Header 2"),
    ("###", "Header 3"),
    ("####", "Header 4"),
    ("#####", "Header 5"),
    ("######", "Header 6"),
]

markdown_split = MarkdownHeaderTextSplitter(headers_to_split_on=headr_to_split_on)

split = markdown_split.split_text(markdown_example)

# Laço para imprimir o conteúdo do documento
for doc in split :
    print(doc.page_content)
    print(doc.metadata)     
    print("==============================================")
```

Split De Documentos

```python
from langchain_community.document_loaders.pdf import PyPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter

chunk_size = 50
chunk_overlap = 10

char_split= RecursiveCharacterTextSplitter(
    chunk_size=chunk_size, 
    chunk_overlap=chunk_overlap,
    separators=['.',' ', ''] #separadores de texto TextSplitter utiliza varias caracteres para dividir .
    
    )
```

```python
caminho ='../arquivos/Explorando o Universo das IAs com Hugging Face.pdf'

loader = PyPDFLoader(caminho)
documentos = loader.load()

splits = char_split.split_documents(documentos)
splits
```
